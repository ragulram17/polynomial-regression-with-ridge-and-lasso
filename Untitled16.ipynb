```
x=df.iloc[:,3:4].values
y=df.iloc[:,8].values
 
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=0)
 
from sklearn.linear_model import LinearRegression
 
from sklearn.model_selection import cross_val_score
LinReg = LinearRegression()
mse=cross_val_score(LinReg,x,y,scoring='neg_mean_squared_error',cv=5)
mean_mse =np.mean(mse)
LinReg.fit(x_train,y_train)
print(mean_mse)
 
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
ridge = Ridge()
parameters={'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20,30,35,40,45,50,55,100]}
ridge_regression =GridSearchCV(ridge, parameters,scoring='neg_mean_squared_error',cv=5)
ridge_regression.fit(x,y)
 
print(ridge_regression.best_params_)
print(ridge_regression.best_score_)
```

```
from sklearn.linear_model import Lasso
lasso = Lasso()
parameters={'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20,30,35,40,45,50,55,100]}
lasso_regression =GridSearchCV(lasso, parameters,scoring='neg_mean_squared_error',cv=5)
lasso_regression.fit(x,y)
 
print(lasso_regression.best_params_)
print(lasso_regression.best_score_)
 
prediction_ridge =ridge_regression.predict(x_test)
prediction_lasso =lasso_regression.predict(x_test)
 
print (prediction_ridge)
print (prediction_lasso)
```

```
import seaborn as sns
import matplotlib.pyplot as plt
sns.distplot(y_test-prediction_ridge)
plt.show()
```

```
sns.distplot(y_test-prediction_lasso)
plt.show()
```

```
#rsquare
 
y_predict_slr = LinReg.predict(x_test)
 
from sklearn import metrics
r_square =metrics.r2_score(y_test,y_predict_slr)
print(r_square)
 
from sklearn import metrics
print('MAE:',metrics.mean_absolute_error(y_test,y_predict_slr))
print('MSE:',metrics.mean_squared_error(y_test,y_predict_slr))
print('RMSE:',np.sqrt(metrics.mean_absolute_error(y_test,y_predict_slr)))
```
